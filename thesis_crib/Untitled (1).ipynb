```python
#Importing necesary data pacakages
import rasterio 
import numpy as np
import matplotlib.pyplot as plt
from skimage import color
from skimage import io
```

```python
img = io.imread('dwellings clipped.tif')
imgGray = color.rgb2gray(img)
plt.imshow(imgGray, cmap="gray")
```

```python
from skimage.filters import roberts, sobel, scharr, prewitt
from skimage.feature import canny

#applying edge detectors
edge_roberts = roberts(imgGray)
edge_sobel = sobel(imgGray)
edge_scharr = scharr(imgGray)
edge_prewitt = prewitt(imgGray)
edge_canny = canny(imgGray, sigma=1)       # it also does noise reduction and gradient

#plotting
plt.figure(figsize=(16,16))
plt.subplot(231)
plt.imshow(imgGray, cmap='gray')
plt.title('original image')
plt.subplot(232)
plt.imshow(edge_roberts, cmap='gray')
plt.title('roberts operator')
plt.subplot(233)
plt.imshow(edge_sobel, cmap='gray')
plt.title('sobel operator')
plt.subplot(234)
plt.imshow(edge_scharr, cmap='gray')
plt.title('scharr operator')
plt.subplot(235)
plt.imshow(edge_prewitt, cmap='gray')
plt.title('prewitt operator')
plt.subplot(236)
plt.imshow(edge_canny, cmap='gray')
plt.title('canny operator')
plt.show()
```

```python

a=np.hstack([np.zeros((256,128)),np.ones((256,128))])
af=np.fft.fftshift(np.fft.fft2(a))
afl=abs(af)
plt.figure()
plt.subplot(121)
io.imshow(a)
plt.subplot(122)
io.imshow(afl)
```

```python
import skimage.exposure as ex
a=np.zeros((256,256))
a[77:177,77:177]=1
af=np.fft.fftshift(np.fft.fft2(a))
afl=ex.rescale_intensity(np.log(1+abs(af)),out_range=(0.0,1.0))

plt.figure()
plt.subplot(121)
io.imshow(a)
plt.subplot(122)
io.imshow(afl)
```
